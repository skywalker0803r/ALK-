```python
#!pip install --upgrade xlrd
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/ALK硫酸腐蝕預測/操作條件.xls',sheet_name='資料擷取')
df
```

```python
c = df.iloc[1,1:].values
v = df.iloc[2:,1:].values
```

```python
d = pd.DataFrame(v,columns=c)
d
```

```python
d = d.dropna(axis=1,how='all')
d = d.dropna(axis=0,how='all')
d = d.fillna(method='ffill')
d = d.fillna(method='bfill')
d.isnull().sum()
```

```python
x_col = d.columns[:-1].values.tolist()
y_col = d.columns[-1]
def lasso_find_coef(df,x_col,y_col):
  from sklearn.linear_model import Lasso
  import pandas as pd
  import numpy as np
  model = Lasso()
  x = df[x_col]
  y = df[y_col]
  model.fit(x,y)
  d = df[x_col].copy()
  d.loc['coef',:] = model.coef_
  return d.loc[['coef'],:]
result = lasso_find_coef(d,x_col,y_col)
```

```python
result.abs().T.sort_values(by='coef',ascending=False)
```
